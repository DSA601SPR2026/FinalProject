```python
import pandas as pd
import os

# Load feature code files from Inputs_Targets folder
base_dir = os.getcwd()
root_dir = os.path.dirname(base_dir)

code_files = ['Risk_Factors', 'Symptoms', 'Targets']

code_lookup = pd.read_csv(os.path.join(root_dir, 'column_name_lookup.csv'))

for i in code_files:
    # Load feature code files from Inputs_Targets folder
    feature_df = pd.read_csv(os.path.join(base_dir, f'{i}.csv'))

    # Map human-readable names
    feature_df['Human-Readable Name'] = feature_df['Variable Code'].map(
        code_lookup.set_index('Column Name')['Human-Readable Name']).fillna('N/A')

    # Map file names from the lookup
    feature_df['File Name'] = feature_df['Variable Code'].map(
        code_lookup.set_index('Column Name')['Data File Name']).fillna('N/A')

    # Reorder columns: first column, then new columns, then the rest
    col1 = feature_df.pop('Human-Readable Name')
    col2 = feature_df.pop('File Name')
    feature_df.insert(1, 'Human-Readable Name', col1)
    feature_df.insert(2, 'File Name', col2)

    # Save updated dataframe back to Inputs_Targets
    feature_df.to_csv(os.path.join(base_dir, f'{i}_Final.csv'), index=False)

    print(f"Preview for {i}:")
    print(feature_df.head())
    print("\n")
```
